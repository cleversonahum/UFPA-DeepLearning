### Setting the Dataset

```python
import numpy as np
from tensorflow import keras
from tensorflow.python.client import device_lib
from math import pi

func = lambda x,y: np.power(np.power((np.power(x,2)*np.power(y,2) + np.power(x,2)*y+1),2),10)

train_set = 60000
test_set = 60000

x = np.random.uniform(-0.5, 0.5, (train_set+test_set,))
y = np.random.uniform(-0.5, 0.5, (train_set+test_set,))

train_x = x[0:train_set]
train_y = y[0:train_set]
train_entries = np.vstack((train_x, train_y)).T
train_z = func(train_x, train_y)

test_x = x[train_set:(train_set+test_set)]
test_y = y[train_set:(train_set+test_set)]
test_entries = np.vstack((test_x, test_y)).T
test_z = func(test_x, test_y)

print(train_z)
print(train_x.shape, "    ", train_y.shape, "     ", train_z.shape, "     ", train_entries.shape)
print(train_x,"\n\n",train_y,"\n\n", train_entries)
print(func(train_x[0], train_y[0]))
```

### Deep NN

```python
model = keras.models.Sequential()

model.add(keras.layers.InputLayer(input_shape=(2,)))
model.add(keras.layers.Dense(9))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation('relu'))
n_hidden = 10
for i in range(n_hidden):
    model.add(keras.layers.Dense(3))
    if(i!=n_hidden-1):
        model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.Activation('relu'))

model.add(keras.layers.Dense(1, activation="linear"))

model.compile(optimizer="adam", loss="mse")

model.summary()
model.fit(train_entries, train_z, batch_size=3000, epochs=2000, verbose=1, shuffle=True)
```

```python
predictions = model.predict(test_entries)
print(predictions.shape, "    ", test_z.shape)
results = np.sqrt(np.mean(np.power(predictions-test_z.reshape((len(test_z),1)),2)))
print(results)
```

### Shallow NN

```python
model = keras.models.Sequential()

model.add(keras.layers.InputLayer(input_shape=(2,)))
model.add(keras.layers.Dense(np.power(2,11)+1))
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.Dense(1, activation="linear"))

model.compile(optimizer="adam", loss="mse")

model.summary()
model.fit(train_entries, train_z, batch_size=3000, epochs=2000, verbose=1, shuffle=True)
```

```python
predictions = model.predict(test_entries)
print(predictions.shape, "    ", test_z.shape)
results = np.sqrt(np.mean(np.power(predictions-test_z.reshape((len(test_z),1)),2)))
print(results)
```

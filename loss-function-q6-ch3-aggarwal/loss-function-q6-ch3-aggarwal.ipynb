```python
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f(x,y):
    return(np.power(x,2)+np.power(y,10))
x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
x,y = np.meshgrid(x,y)
z = f(x,y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
fig.show()
```

```python
lr = 0.001 #Learning rate
N = 10 #Number of steps

# First Case
x1 = 0.5
y1 = 0.5
z_arr1 = np.zeros([N])

# Second case
x2 = 2
y2 = 2
z_arr2 = np.zeros([N])

for i in range(N):
    #Fisrt case
    partial_der_x1 = 2*x1
    partial_der_y1 = 10*np.power(y1,9)
    x1 = x1 - lr*partial_der_x1
    y1 = y1 - lr*partial_der_y1
    z_arr1[i] = f(x1,y1)
    
    #Second case
    partial_der_x2 = 2*x2
    partial_der_y2 = 10*np.power(y2,9)
    x2 = x2 - lr*partial_der_x2
    y2 = y2 - lr*partial_der_y2
    z_arr2[i] = f(x2,y2)
print(z_arr1, z_arr2)
```

```python
plt.figure()
plt.plot(range(N), z_arr1, 'or', label="x=0.5 y=0.5")
plt.plot(range(N), z_arr2, '+b', label="x=2 y=2")
plt.plot(range(N), np.abs(z_arr1-z_arr2), '-g', label="Diff")
plt.legend()
plt.show()
```
